```python
import requests
from bs4 import BeautifulSoup
import textwrap
import re
```

```python
# Set up headers to mimic a browser visit
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Referer": "https://google.com"
}


# Fetch and parse the webpage
url = 'https://understandingwar.org/research/russia-ukraine/russian-offensive-campaign-assessment-november-4-2025/'
response = requests.get(url, headers=headers)
html = response.text

# Parse with BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract title
article_title = soup.title.string
title = article_title.strip()
print(title)
```

```python
# Step 3: Remove unwanted tags
for tag in soup(["script", "style", "noscript", "header", "footer", "nav", "aside"]):
    tag.decompose()

# Step 4: Remove paragraphs that contain <a> tags
clean_paragraphs = [
    p.get_text(strip=True)
    for p in soup.find_all("p")
    if not p.find("a")
]

# Step 5: Combine paragraphs
clean_text = "\n\n".join(clean_paragraphs)

# Step 6: Remove URLs and website references
clean_text = re.sub(r"http\S+|www\.\S+|https\S+", "", clean_text)

# Step 7: Remove [number] references
clean_text = re.sub(r"\[\d+\]", "", clean_text)

# Step 8: Remove everything after "WARNING:"
clean_text = re.split(r"WARNING:", clean_text, flags=re.IGNORECASE)[0]

# Step 9: Clean up extra whitespace
clean_text = re.sub(r"\s{2,}", " ", clean_text).strip()

# Step 10: Print result
clean_text = textwrap.fill(clean_text, width=180)
print(clean_text)
```
